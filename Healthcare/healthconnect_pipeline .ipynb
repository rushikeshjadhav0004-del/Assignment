```python
import pandas as pd



patients_df = pd.read_csv(r"C:\Users\jsuya\Desktop\Data_1\synthea_data\patients.csv")
encounters_df = pd.read_csv(r"C:\Users\jsuya\Desktop\Data_1\synthea_data\encounters.csv")
conditions_df  = pd.read_csv(r"C:\Users\jsuya\Desktop\Data_1\synthea_data\conditions.csv")
observations_df = pd.read_csv(r"C:\Users\jsuya\Desktop\Data_1\synthea_data\observations.csv")
```

```python
# Preview encounters table
print("Encounters Table Preview:")
encounters_df.head(3)
```

```python
# Preview conditions table
print("Conditions Table Preview:")
conditions_df.head(3)
```

```python
import hashlib

def hash_value(val):
    """Takes any value and returns a short SHA-256 hash string."""
    if pd.isnull(val):
        return None
    return hashlib.sha256(str(val).encode()).hexdigest()[:16]  # first 16 chars only

pii_columns = ["FIRST","MIDDLE" "LAST", "SSN", "ADDRESS"]


patients_masked = patients_df.copy()

patients_masked["HASHED_ID"] = patients_masked["Id"].apply(hash_value)

# Drop only the PII columns that actually exist
cols_to_drop = [c for c in pii_columns if c in patients_masked.columns]
patients_masked.drop(columns=cols_to_drop, inplace=True)

print("PII columns removed:", cols_to_drop)
patients_masked.head(3)
```

```python

encounters_df["HASHED_ID"]   = encounters_df["PATIENT"].apply(hash_value)
conditions_df["HASHED_ID"]   = conditions_df["PATIENT"].apply(hash_value)
observations_df["HASHED_ID"] = observations_df["PATIENT"].apply(hash_value)

print("HASHED_ID added to Encounters, Conditions, Observations tables!")
print("Sample hashed IDs from encounters:")
encounters_df[["PATIENT","HASHED_ID"]].drop_duplicates().head(4)
```

```python
total_cost = (
    encounters_df
    .groupby("HASHED_ID")["TOTAL_CLAIM_COST"]
    .sum()
    .reset_index()
    .rename(columns={"TOTAL_CLAIM_COST": "TOTAL_ENCOUNTER_COST"})
)

print("Total Encounter Cost per patient sample:")
total_cost.head()
```

```python
conditions_df["START"] = pd.to_datetime(conditions_df["START"], errors="coerce")

latest_condition = (
    conditions_df
    .sort_values("START", ascending=False)
    .drop_duplicates(subset="HASHED_ID", keep="first")
    [["HASHED_ID", "DESCRIPTION", "START"]]
    .rename(columns={
        "DESCRIPTION": "LATEST_CONDITION",
        "START": "CONDITION_DATE"
    })
)

print("Latest Condition per patient (sample):")
latest_condition.head()
```

```python
master_view = patients_masked.copy()

master_view["HASHED_ID"] = master_view["Id"].apply(hash_value)
master_view = master_view.merge(total_cost, on="HASHED_ID", how="left")
master_view = master_view.merge(latest_condition, on="HASHED_ID", how="left")
master_view.drop(columns=["Id"], inplace=True)

master_view.head(5)
```

```python
master_view.to_csv("output.csv", index=False)
master_view.head(5)
```
